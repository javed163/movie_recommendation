```python
popularity_df = num_rating_df.merge(avg_rating_df, on = 'Book-Title')
```

```python
popular_df = popularity_df[popularity_df['num_rating']>=250].sort_values('avg_rating',ascending = False).head(50)
```

# Collaborative Filtering Based Recommender System

```python
x = rating_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_lekhe_user = x[x].index
```

```python
filtered_rating = rating_with_name[rating_with_name['User-ID'].isin(padhe_lekhe_user)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating.drop_duplicates()
```

```python
pt = final_rating.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt).shape
```

```python
similarity_scores = cosine_similarity(pt)
shape_of_similarity_scores = similarity_scores.shape
print(shape_of_similarity_scores)
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:6]

    #similar_items = similarity_scores[index]
    data = []
    for i in similar_items:
        item = []

        temp_df = book[book['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        # item.append(temp_df.drop_duplicates('Book-Title')[''])

        data.append(item)

    return data
```

```python
recommend('The Da Vinci Code')
```
